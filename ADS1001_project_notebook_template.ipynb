```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("melb_data.csv")

df.head()
df.info()
```

```python
- The dataset was loaded from a CSV file (melb_data.csv) using pd.read_csv().

- the dataset describes dwelling Melbourne housing sales between 2016 - 2017 and contains information on property sales 
incuding price, number of rooms, distance from the CBD, suburb, and other property features.
    
- The dataset consists of approximately 13,580 observations and 21 variables
with a mix of numerical (e.g., Price, Rooms, Distance) and categorical (e.g., Suburb, Type) data.
    
- Initial inspection using df.info() showed that most key variables such as Price, Rooms, and Distance had no missing values,
while some variables (e.g., BuildingArea and YearBuilt) contained missing data.
    
```

## 6. Data Cleaning

Explain how missing values, duplicates, and errors are handled. Include reasoning for any deletions, imputations, or corrections.

```python
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds (standard multiplier is 1.5)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the data
df_clean = df[(df['Price'] >= lower_bound) & (df['Price'] <= upper_bound)]
```

- Outliers in the Price variable were identified and removed using the Interquartile Range (IQR) method. 
    The first quartile (Q1) and third quartile (Q3) were calculated, and the IQR was defined as Q3 − Q1.
    Any values below Q1 − 1.5×IQR or above Q3 + 1.5×IQR were considered outliers and removed.

```python
# plot raw data (before removing outliers)
plt.figure(figsize=(10, 8))

sns.scatterplot(
    data=df,                 # original dataset
    x='Distance',            # distance from CBD
    y='Price',               # house price
    hue='Type',              # colour by property type
    alpha=0.7                # transparency so points don’t overlap too much
)

plt.title('Price vs Distance (Before Cleaning)')
plt.xlabel('Distance from CBD (km)')
plt.ylabel('Price ($)')

plt.show()
```

```python
# plot cleaned data (after removing outliers)
plt.figure(figsize=(10, 8))

sns.scatterplot(
    data=df_clean,           # cleaned dataset (outliers removed)
    x='Distance',
    y='Price',
    hue='Type',
    alpha=0.7
)

plt.title('Price vs Distance (After Removing Outliers)')
plt.xlabel('Distance from CBD (km)')
plt.ylabel('Price ($)')

plt.show()
```

```python
After removing outliers, the data becomes less scattered and the relationship between price and 
distance is clearer and easier to interpret.
```
