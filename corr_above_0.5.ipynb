```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import PolynomialFeatures, StandardScaler

df = pd.read_csv('cleaned_melb_dataset.csv')
print(df.shape)
```

```python
df_encoded = pd.get_dummies(
    df, columns=["Type", "Regionname"], drop_first=False, dtype=int
)

# 2. Calculate correlation matrix specifically against Price
price_corr = df_encoded.corr(numeric_only=True)["Price"]

# 2. Automatically find all columns where the absolute correlation is 0.5 or higher
features_05 = price_corr[abs(price_corr) >= 0.5].index.tolist()

# 3. Automatically slice your dataframe using that list
df_above_05 = df_encoded[features_05]
```

```python
# Print the list to verify what Python selected for you
print("Automatically selected features for 0.5:", df_above_05.columns.tolist())
```

```python
# 4. Save to CSV without manual typing
df_above_05.to_csv("housing_above_05_automated.csv", index=False)
```
